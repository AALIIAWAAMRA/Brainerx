# TCP Server Connections with nc (Netcat)  

Step 1: The Python TCP Server
Below is the python code to initialize a TCP server. It will listen on 127.0.0.1 (localhost) at port 65432. Once a client connects, it greets them with a success message and terminates the connection cleanly.

```python
import socket

HOST = '127.0.0.1'  
PORT = 65432        

with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as server_socket:
      server_socket.bind((HOST, PORT))
      server_socket.listen(1)
      print(f"[INFO] Server is running and listening on {HOST}:{PORT}...")

      conn, addr = server_socket.accept()
      with conn:
         print(f"[SUCCESS] Connected successfully by: {addr}")
         
         message = "Welcome! You have successfully connected to the TCP server.\n"
         conn.sendall(message.encode('utf-8'))
```

Step 2: Connecting with Netcat (nc)
To test and interact with our running TCP server, we use Netcat, often referred to as the Swiss Army knife of networking.

Open your terminal (or command prompt) and run the following command while the script above is running:

```python
nc 127.0.0.1 65432
```
